##Aim

>To plot different signals and analyze their PDF and CDF




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Cdf(y):
    count,bin_count=np.histogram(y,bins=10)
    pdf=count/sum(count)
    cdf=np.cumsum(pdf)
    plt.figure();
    plt.plot(bin_count[1:],cdf)
```

```python
x1=np.random.uniform(0,1,10000)
plt.hist(x1);
Cdf(x1)
```

```python
x2=np.random.exponential(1,1000)
plt.hist(x2);
Cdf(x2)
```

```python
x3=np.random.normal(0.0,5,1000)
plt.hist(x3);
Cdf(x3)
```

```python
x4=np.random.rayleigh(10,1000)
plt.hist(x4);
Cdf(x4)
```
